# Multiple regression model

### Steps

  1. Identify x (IVs) and y (DV): x=Height,Width,VerticalLength, y=Weight
  2. Create Train and Test datasets
  3. Train model
  4. Evaluate model

```python
# Focusing on Bream
# query() used to select all data where Species == Bream
bream = fish.query('Species == "Bream"')
```

```python
# Identify IVs and DV
x = bream[['Height','Width','VerticalLength']]
y = bream[['Weight']]
```

```python
# Create train and test datasets
# a. split variables into training and testing sets
# b. build model using training, then run on testing

# Train/Test split 75%/25%
x_train, x_test, y_train, y_test = train_test_split(x, y, train_size = 0.75, test_size = 0.25, random_state = 100)
```

```python
x_train.head(3)
```

```python
# Display y training data converted from grams to LBs
(y_train/454).head(3)
```

## Fitting model: Best fitting line

```python
# Using LinearRegression()
model = LinearRegression().fit(x_train, y_train)
```

```python
# Display R2 correlation value
# score() function: Accepts test dataset and returns R2 correlation value
# Scale of 0-1: higher score indicates a better fit
model.score(x_test, y_test)
```

```python
# Predict weight and display predicted values based upon IVs
y_predicted = model.predict(x_test)

# Create predicted dataframe with "PredictedWeight" col using y_predicted values
predicted = pd.DataFrame(y_predicted, columns = ['PredictedWeight'])
predicted
```

```python
# THESE WEIGHTS ARE IN GRAMS, A BREAM WILL NOT WEIGH 1000LBS
# Join and display all cols, including PredictedWeight col
# drop parameter removes old index vals being added as col
final = predicted.join([x_test.reset_index(drop=True),
                        y_test.reset_index(drop=True)])

# Display predictedweight and actual weight (DVs) and IVs
final
```

<br>

## Residuals

```python
# "Residuals" are simply differences between DV test values and DV predicted values
# Residual = Actual - Predicted
final['Residual'] = final.Weight - final.PredictedWeight
final
```

```python
# Kernel Density Estimate (KDE) plot: Estimates data distribution
# Visual representation of underlying distribution of data
# Produces small continous curve (also called kernel) for every individual data point along an axis

sns.kdeplot(data=final, x='Residual')
```

**Based on the KDE plot, we can see our findings are centered near 0, showcasing that the predictions are close to the actual values. It can also be noted that the KDE plot has a long left tail, which would highlight the cause of overpredicting cases (prediction weight vastly greater than actual weight) seen above.**
